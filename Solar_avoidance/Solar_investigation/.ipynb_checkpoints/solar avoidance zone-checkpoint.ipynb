```python
import os
import katdal
import katpoint
%pylab inline
import scape
import string
import time
%matplotlib inline
```

```python
name = '1657625710' #'1607433065'  # e.g 1583650035
file_=str('http://archive-gw-1.kat.ac.za:7480/'+name+'/'+name+'_sdp_l0.full.rdb');
rdb = katdal.open(file_)
```

```python
def data_cal(file, nd_models):
    ants = file.ants
    for ant in range(len(ants)):
        
        data = scape.DataSet(file, baseline="%s" %(ant.name), nd_models=nd_models)
        data = data.convert_power_to_temperature()
        data = data.select(flagkeep='~nd_on') #removing the noise diode
        data = data.select(labelkeep='scan') #keeps the scan and remove anything at the end
    return data
```

```python
ants = rdb.ants
for index, ant in enumerate(ants):
    data = scape.DataSet(rdb, baseline="%s" %(ant[index].name), nd_models=nd_models)
print(data)
```
