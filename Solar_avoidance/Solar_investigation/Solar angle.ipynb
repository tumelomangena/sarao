```python
def counting_bad_ants(ffile):
    '''This function checks the quality of data by determining 
    how many antennas stopped, or stuck during the observation.
    
    Input: Take a fullvis rdb file from the archive
    Output: The number of bad antennas, The list of bad antennas'''
  
    
    #This pull all the antenna used for observation
    AntList = []
    for ant in ffile.ants:
        AntList.append(ant.name)

    #This will give the antenna activity list
    AntsActivity = []
    for AntName in AntList:
        AntsActivity.append((AntName, ffile.sensor[AntName+'_activity']))
    
    #Identifies and count number of bad atennnas
    badants = []
    for i in range(len(AntsActivity)):
        ants_list = AntsActivity[i][1].tolist()
        counting = ants_list.count('stop')
        if counting > 15:
            badants.append(AntsActivity[i][0])
        else:
            pass
        
    #Calculating the percentage of the quality of data based on number of bad antennas    
    perc = (float(len(badants))/float(len(AntList)))*100
    
    
    if len(badants) == 0:
        print("No bad antennas found.")
    else:
        print('%s antennas considered bad' % len(badants))
        print('Bad antennas is/are %s' % badants)
    #print 'The quality of data based on number of bad antennas is %s' % (100 - perc)

    return 'Bad antennas is/are %s' % badants, badants
```
